# ANN CLASSIFICATION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.drop(['RowNumber', 'CustomerId', 'Surname'],axis=1)
df.head()
```

```python
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df['Gender'])
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe_geo=OneHotEncoder()
geo_encoder=ohe_geo.fit_transform(df[['Geography']])
geo_encoder
```

```python
ohe_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=ohe_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoder_df
```

```python
df=pd.concat([df.drop('Geography',axis=1),geo_encoder_df],axis=1)
df.head()
```

```python
df.head()
```

```python
X=df.drop(columns='Exited')
y=df['Exited']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X
```

```python
y
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

# ANN IMPLIMENTATION

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train_scaled = X_train_scaled[:, :11]
X_test_scaled = X_test_scaled[:, :11]
```

->Define hidden layers

```python
model = Sequential()

model.add(Dense(11,activation='relu',input_dim=11))
model.add(Dense(13,activation='relu'))
model.add(Dense(13,activation='relu'))
model.add(Dense(13,activation='relu'))
model.add(Dense(13,activation='relu'))
model.add(Dense(13,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

After doing this we have to do comple in which we define

->Loss function
->optimizers


```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False
)
```

```python
model.fit(X_train_scaled,y_train,epochs=10000,validation_split=0.2,callbacks=[callback])
```

```python
model.layers[0].get_weights()
```

```python
model.predict(X_test_scaled)
```

->Now we set up the prediction as :

->yes:1
->No:0

```python
y_log = model.predict(X_test_scaled)
y_pred = (y_log > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
